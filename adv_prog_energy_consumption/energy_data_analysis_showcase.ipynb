## Historical evolution of GDP and Total Energy Consumption
In the following section we compare the global energy consumption, broken down by country in a GapMinder style. Each dot represents a country and its dimension is proportional to the population of the country.
We compare two years: 1990 and 2016. The country observed is highlighted in orange.

We then plot the yearly consumption and the GDP for the three countries. It's remarkable to notice that before 2010 China overtook the US in terms of total energy consumption and managed to almost equal the aggregate GDP level of the US.

```python
Energy_data.gapminder(year = 1990, country = "China")
```

```python
Energy_data.gapminder(year = 2016, country = "China")
```

Lead by energy-intensive activities, mainly manufacturing and in particular steelmaking, over less than two decades China developed massively but the rapid economic growth was paired with substantial increase in enegry consumption.

```python
Energy_data.gapminder(year = 1990, country = "United States")
```

```python
Energy_data.gapminder(year = 2016, country = "United States")
```

The United States followed a similar but less abrupt trajectory. While already, and still, the first economy of the world in terms of GDP, it maintained its position but the growth in energy consumption, while still sustained, remained inferior to that of China.

```python
Energy_data.gapminder(year = 1990, country = "Portugal")
```

```python
Energy_data.gapminder(year = 2016, country = "Portugal")
```

The evolution of Portugal was different. The country enjoyed moderate and intermittent growth in terms of GDP paired with an equally moderate gorwth in energy consumption. In general, it lies with other countries in an advanced state of development.

## Emissions

Now turning to emissions we plot the total amount of $CO_{2}$ emitted by each country against the country population. It's noticeable that there are large-population countries whose level of emission is proportional to the energy consumption. Above the $45°$ line that bisects our plot, lies a populous cluster of countries that are smaller in terms of population but whose energy consumption is not in line with the emissions.

## Predictions

We finally employ autoregressive analysis to make predictions about the possible future developments of $CO_{2}$ emissions. 
Thanks to its already substantial and growing renewable capacity, predictions for Portugal indicate a stable level of energy consumption paired with a reduction in greenhouse gas emissions. On the other hand China is expected to continue growing economically and in terms of energy consumption but its energy mix is likely to cause the predicted further increase in $CO_{2}$ emissions.

```python
Energy_data.prediction("China", periods = 10)
```

## Conclusions

Factors played different roles in various areas. Generally, coal consumption intensity accounted most in the change of the  energy consumption, which is a serious question in China. The use of coal caused the growth of energy consumption but it is characterized by of high pollution. As a consequence, the adjustment of energy mix is in desperate need. China is developing with a fast speed and more energy has been put into use than ever before.
For developing countries like China, the dependence of coal consumption is primarily for the reason of its low price however, in the long run, government can link the switch to a more sustainable energy consumption development pattern and environment friendly society with offering more job opportunities.
When it comes to the USA, the most significant factor is the GDP. In other words, economic growth was the greatest contributor to energy use. As the world's first largest economy, the USA always consumes a large amount of energy to satisfy the need to increase the domestic product. In other ways, human activities, especially in terms of the economic aspect, have a strong connection with energy use to maintain industrial and daily life. Moreover, the USA consume an enormous amount of energy produced from oil.
Wavering committments to Paris Agreements from the US outline an uncertain path ahead.
Portugal offers an interesteing example: it's heterogenous energy mix proved to be an interesting solution and put the country on a predicted path of emissions reduction.
In any case, for all the countries analysed the government needs to pour more funds to these sustainable industries to speed up the process of building a more sustainable energy system with lower environment impacts.
